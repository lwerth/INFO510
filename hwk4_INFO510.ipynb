> the ordinary linear regression model from 15.1 (a), ignores the affects each treatment has on each other and treats them as 16 independent instead of 3 sets of temperature variables with about 5 observations in each. Because each treatment is seen as indepentent, the model is prone to overfitting. Using a hierarchcal variance parameter allows correlation within the explanatory variables, and is less prone to positive extremes (overfitting). Hierarchcal variance also shows different components of variability for each temperature which is indicative of stronger reliability.
