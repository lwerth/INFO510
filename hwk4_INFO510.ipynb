15.3 (c) Discuss the differences between the inferences in (a) and (b). Interpret the differences in terms of the hierarchical variance parameter. Do you agree with Marquardt and Snee that the inferences from (a) are unacceptable?

> The ordinary linear regression model from 15.1 (a), ignores the affects each treatment has on each other and treats them as 16 independent variables showing a a shifting Posterior distribution. Because each treatment is seen as indepentent, the model is prone to overfitting, which is why the distribution fluctuates, it follows each prior. Using a hierarchcal variance parameter allows correlation within the explanatory variables, and shows a normal distribution when sampling the posterior. It is less prone to positive extremes (overfitting). Hierarchcal variance also shows different components of variability for each temperature which is indicative of stronger reliability.
